```python
import cv2
from google.colab.patches import cv2_imshow
from PIL import Image
import numpy as np

file_name = 'Your FileName'

# Making Video
cap = cv2.VideoCapture(file_name)
video_output_path = '/content/drive/MyDrive/detect_result.mp4'
# ì €ì¥í•  ë¹„ë””ì˜¤ íŒŒì¼ì˜ ì½”ë± 
codec = cv2.VideoWriter_fourcc(*'XVID')

#ë¹„ë””ì˜¤ í™”ë©´ í¬ê¸°
#cv2.CAP_PROP_FRAME_WIDTH : file_name ë¹„ë””ì˜¤ í”„ë ˆì„ì˜ ê°€ë¡œ
#cv2.CAP_PROP_FRAME_HEIGHT) : file_name ë¹„ë””ì˜¤ í”„ë ˆì„ì˜ ì„¸ë¡œ
vid_size = (round(cap.get(cv2.CAP_PROP_FRAME_WIDTH)),round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))

#file_name ë¹„ë””ì˜¤ì˜ ì´ˆë‹¹ í”„ë ˆì„
vid_fps = cap.get(cv2.CAP_PROP_FPS)

#cv2.VideoWriter : ë¹„ë””ì˜¤ íŒŒì¼ì„ ìƒì„±í•´ì„œ ì €ì¥í•  ê°ì²´ ìƒì„±
#cv2.VideoWriter(ì €ì¥í•  ê²½ë¡œ, ë¹„ë””ì˜¤ì½”ë±, ë¹„ë””ì˜¤ ì´ˆë‹¹ í”„ë ˆì„, í”„ë ˆì„í¬ê¸°)    
vid_writer = cv2.VideoWriter(video_output_path, codec, vid_fps, vid_size) 
```
